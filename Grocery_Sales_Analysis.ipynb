##### Monthly Orders

```python
monthly_order = df_sales.groupby('SalesDate').agg({'TotalPrice':'sum','SalesID':'nunique'})
```

```python
monthly_order.reset_index()
monthly_order.rename(columns={'SalesDate':'month','TotalPrice':'Total_Sales','SalesID':'OrdersCounts'},inplace=True)
monthly_order.to_csv('monthly_order.csv')
```

##### Monthly Customers

```python
df_sales['month'] = df_sales['SalesDate'].astype(str).apply(lambda x: x.split('-')[1])
monthly_cust = df_sales.groupby('month').agg({'CustomerID':'nunique'})

monthly_cust = monthly_cust.reset_index()
monthly_cust.rename(columns={'month':'month','CustomerID':'CustomersCounts'},inplace=True)

monthly_cust['month'] = monthly_cust['month'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%B'))
monthly_cust.to_csv('monthly_cust.csv')
```

##### Staff performance by their service period


---



```python
emp = df_sales.merge(df_employee,on='SalesPersonID')
```

```python
emp['HireDate'] = pd.to_datetime(emp['HireDate'])

# Find the earliest and latest hire dates in the dataset
start_date = emp['HireDate'].min()
latest_date = emp['HireDate'].max()

# Calculate the midpoint (half period)
half_period = start_date + (latest_date - start_date) / 2

# Classify staff based on hire_date compared to the midpointciy
emp['ranked'] = emp['HireDate'].apply(lambda x: 'new staff' if x > half_period else 'old staff')

# Add midpoint for reference
emp['half_period'] = half_period

emp.head()
```

```python
emp['HiredDate'] = emp['HireDate'].astype(str).apply(lambda x: x.split(' ')[0])
```

```python
emp['HiredDate']
```

```python
staff_performances  = emp.groupby(['ranked','SalesPersonID','month']).agg({'TotalPrice':'sum','SalesID':'nunique','CustomerID':'nunique'})
```

```python
staff_performances.rename(columns={'month':'month','SalesPersonID':'SalesPerson','TotalPrice':'Total_Sales','SalesID':'OrdersCounts','CustomerID':'CustomersCounts'},inplace=True)
staff_performances.to_csv('staff_performances.csv')
```

```python
deep  = emp.groupby(['HiredDate','month','ranked','SalesPersonID']).agg({'TotalPrice':'sum','SalesID':'nunique','CustomerID':'nunique'})
```

```python
deep.rename(columns ={'HiredDate': 'StartDate','SalesPersonID':'SalesPerson','TotalPrice':'Total_Sales','SalesID':'OrdersCounts','CustomerID':'CustomersCounts'},inplace=True)
deep.reset_index()
```
